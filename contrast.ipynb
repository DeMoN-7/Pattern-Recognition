Increasing contrast using histogram equalization

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("muki.jpg",0)
equalized=cv2.equalizeHist(img)
```

```python
plt.imshow(img,cmap='gray')
```

```python
plt.imshow(equalized,cmap='gray')
```

Contrast stretching

```python
min_val = np.min(img)
max_val = np.max(img)
stretched = (img - min_val) * (255 / (max_val - min_val))
```

```python
plt.imshow(stretched,cmap='gray')
```

```python
noise = np.random.normal(0, 25, img.shape).astype(np.uint8)  # Generate noise
noisy_image = cv2.add(img, noise)
```

```python
plt.imshow(noisy_image,cmap='gray')
```
