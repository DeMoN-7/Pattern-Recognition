```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

# Load and preprocess the image
image = cv2.imread('muki.jpg', cv2.IMREAD_GRAYSCALE)  # Read image in grayscale
image = cv2.resize(image, (256, 256))  # Resize for better visualization
```

```python
# Perform 2D Discrete Wavelet Transform (DWT)
wavelet = 'haar'  # You can change this to 'db1', 'coif1', etc.
coeffs2 = pywt.dwt2(image, wavelet)
```

```python
# Extract approximation and details
cA, (cH, cV, cD) = coeffs2  # Approximation, Horizontal, Vertical, Diagonal details
```

```python
# Reconstruct the image using inverse DWT
reconstructed = pywt.idwt2(coeffs2, wavelet)
```

```python

# Plot the images
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(cA, cmap='gray')
plt.title("Approximation (Low-Frequency)")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(cH, cmap='gray')
plt.title("Horizontal Detail")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(cV, cmap='gray')
plt.title("Vertical Detail")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(cD, cmap='gray')
plt.title("Diagonal Detail")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(reconstructed, cmap='gray')
plt.title("Reconstructed Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
